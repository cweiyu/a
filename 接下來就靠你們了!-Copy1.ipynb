```python
edges_list = []
for i in title_list:
    comb = combinations(i, 2)
    for j in list(comb): 
        edges_list.append(j)
```

```python
G = nx.Graph()
G.add_edges_from(edges_list)
G.remove_edges_from(G.selfloop_edges())
```

```python
# Compute the degree of every node: degrees
degrees = []
for n in G.nodes():
    degree = len(list(G.neighbors(n)))
    degrees.append(degree)
# Print the degrees
print(degrees)
```

```python
def scanfriends(x,n):
    E = 0
    total = []
    near = []
    friend = []
    near.append(inf['neighbor'][x])
    near = near[0]
    while n > 1:
        new = []
        for i in near:
            for a in inf['neighbor'][i]:
                new.append(a)
        new = list(set(new))
        n -= 1
        friend = near + new
    friend += near
    friend.append(x)
    friend = list(set(friend))    
        
    for j in friend:
        for k in inf['neighbor'][j]:
            if k in friend and k != j:
                #print (j,k)
                E += 1
    Edge = E / 2
    
    total = (len(friend) * (len(friend) - 1)) / 2
    if total > 0:
        afa = Edge / total
        return afa,sorted(friend)
    else:
        return 'none','none'
    

    #comb2 = list(itertools.combinations(list(G.nodes()),2))
    #Beta = len(list(G.edges()))/len(comb2)
    #if afa > bata:
    #    LR = (afa ** Edge) * (bata ** )
    #else:
    #    LR = 1
   
```

```python
def FDC(x,n):
    f = 0
    while n >0:
        data = ['class1','class2','class3','class4','class5','class6']
        ave = 0
        if str(final[data[n-1]][node.index(x)]) == 'nan':
            return np.NaN
        else:    
            for i in final[data[n-1]][node.index(x)]:
                ave += degrees[node.index(i)]
            ave = ave / (len(final[data[n-1]][node.index(x)]))
            f += math.exp(-(n**(2))) * (degrees[node.index(x)] - ave)
            n -= 1

    return f
```

```python
def overlap(x):
    drop = []
    for i in x:
        #print(i)
        if len(i) > 1: 
            n = 0
            for j in range(n+1,len(i)):
                #print(i[j])
                over = 0
                for k in i[n]:
                    #print(k)
                    if k in i[j]:
                        over += 1
                if over/max(len(i[n]),len(i[j])) >= 0.05:
                    if len(i[n]) < len(i[j]):
                        drop.append(i[n])
                        n += 1
                    else:
                        drop.append(i[j])
    return drop 
```

```python
def dropp(x):
    a = overlap(x)
    b = []
    c = []
    for i in a:
        b.append(sorted(i))
    b = set(tuple(b))
    for i in b:
        c.append(list(i))
    return c
```
